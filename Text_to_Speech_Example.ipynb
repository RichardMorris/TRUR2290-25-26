## Text to Speech Model Example

From the publicly accessible model [***Kokoro-82m***](https://huggingface.co/hexgrad/Kokoro-82M). Not all models work directly with the *Transformers* package.

In this example, the Kokoro model is imported through it's own library. This is done for speed as while the functions used are similar, the default methods are not optimised for this application. What Hugging Face provides is an easy jumping off point for anyone to start using AI today.

### Setup

```python
# 3️⃣ Initalize a pipeline
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
```

### Input

```python
VOICE = "bf_emma" # @param ["af_heart", "af_bella", "af_nicole", "af_aoede", "am_fenrir", "bf_emma", "bm_fable"]
```

```python
text = "A question many of us are asking these days, What is AI? Artificial Intelligence. A name that brings to mind robots and futuristic machines, but AI is both older and simpler than you might think. " #@param {type:"string"}
```

### Run The Model

```python
pipeline = KPipeline(lang_code=VOICE[0])

# 4️⃣ Generate, display, and save audio files in a loop.
generator = pipeline(
    text,
    voice=VOICE,
    speed=0.9, split_pattern=r'\n+'
)
for i, (gs, ps, audio) in enumerate(generator):
    print(gs) # gs => graphemes/text
    print(ps) # ps => phonemes
    display(Audio(data=audio, rate=24000, autoplay=i==0))
    sf.write(f'{i}.wav', audio, 24000) # save each audio files
```
